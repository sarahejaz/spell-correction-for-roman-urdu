**Substitution Table**

````python
def make_sub_table(words_dict):
  sub_table = [[0 for x in range(27)] for y in range(27)]

  for word in words_dict:
    word_list = words_dict[word]
    word = '@' + word
    for i in range(0, len(word_list)):
      miss_word = word_list[i]
      miss_word = '@' + miss_word
      if (len(word)==len(miss_word)): #substitute and transpose
        for j in range(0, len(miss_word)):
          letter1 = miss_word[j]
          if (letter1 not in word): #substitute
            for k in range(0, len(word)):
              if (miss_word[k] != word[k]):
                row = ord(miss_word[k]) - 96
                col = ord(word[k]) - 96
                sub_table[row][col] += 1

  return sub_table
````

**Transpose Table**

````python
def make_trans_table(words_dict):
  trans_table = [[0 for x in range(27)] for y in range(27)]

  for word in words_dict:
    word_list = words_dict[word]
    word = '@' + word
    for i in range(0, len(word_list)):
      miss_word = word_list[i]
      miss_word = '@' + miss_word
      if (len(word)==len(miss_word)): #substitute and transpose
        trans_len = 0
        for k in range(0, len(miss_word)):
          if miss_word[k]==word[k]:
            trans_len += 1

        if (trans_len)==(len(word)-2):
          for k in range(0, len(word)-1):
            if (word[k]==miss_word[k+1] and word[k+1]==miss_word[k]) and (word[k]!=miss_word[k] and word[k+1]!=miss_word[k+1]):
              row = ord(word[k]) - 96
              col = ord(miss_word[k]) - 96
              trans_table[row][col] += 1

  return trans_table
````

**Insert Table**

````python
def make_ins_table(words_dict):
  ins_table = [[0 for x in range(27)] for y in range(27)]

  for word in words_dict:
    word_list = words_dict[word]
    word = '@' + word + '#'
    for i in range(0, len(word_list)):
      miss_word = word_list[i]
      miss_word = '@' + miss_word
      if (len(miss_word)==len(word)): #insert
        for k in range(0, len(miss_word)):
          if (miss_word[k]!=word[k]):
            if (miss_word[k-1]=='@' and word[k-1]=='@'):
              row = 0
              col = 0
            else:
              row = ord(word[k-1]) - 96
              col = ord(miss_word[k]) - 96
            ins_table[row][col] += 1


  return ins_table
````

**Delete Table**

````python
def make_del_table(words_dict):
  del_table = [[0 for x in range(27)] for y in range(27)]

  for word in words_dict:
    word_list = words_dict[word]
    word = '@' + word
    for i in range(0, len(word_list)):
      miss_word = word_list[i]
      miss_word = '@' + miss_word + '#'
      if (len(miss_word)==len(word)): #delete
        for k in range(0, len(word)):
          if (miss_word[k]!=word[k]):
            #print(word, miss_word)
            if (miss_word[k-1]=='@'):
              row = 0
              col = ord(miss_word[k]) - 96
            else:
              row = ord(miss_word[k-1]) - 96
              col = ord(word[k]) - 96
            del_table[row][col] += 1

  return del_table
````

# Calculating Probabilities P(x|w)

**Character-Level Bigram Model**

````python
def generate_bigrams(words_list):
  bigrams = []
  b = ""
  for w in words_list:
    temp_w = '@' + w + '#'
    for i in range(0, len(temp_w)-1):
      b += temp_w[i]
      b += temp_w[i+1]
      bigrams.append(b)
      b = ""

  return bigrams
````

````python
def bigram_model(bigrams):
  freq = {}
  for l in bigrams:
    if not l in freq:
      freq[l] = 1
    else:
      freq[l] += 1

  #print(freq)
  return freq
````

**Finding Candidate Words**

````python
def find_candidate_words(all_words, correct_words, miss_word):
  miss_word = miss_word.lower()
  candidate_words = []
  for w in correct_words:
    if len(miss_word)==len(w): #substitute and transpose
      sub_len = 0
      trans_len = 0
      for i in range(0, len(miss_word)):
        if miss_word[i]==w[i]:
          sub_len += 1
          trans_len += 1

      if (sub_len)==(len(w)-1):
        if (w in all_words):
          candidate_words.append(w)

      if (trans_len)==(len(w)-2):
        trans_flag = False
        i = 0
        while (i<len(miss_word)-1 and trans_flag == False):
          if (w[i]!=miss_word[i]):
            #print(w[i], miss_word[i])
            if ((w[i]==miss_word[i+1]) and (w[i+1]==miss_word[i])):
              #print(w, miss_word)
              trans_flag = True
          i+=1

        if trans_flag == True:
          if (w in all_words):
            candidate_words.append(w)
      
    if (len(miss_word)==(len(w)+1)): #insert
      ins_len = 0
      i = 0
      j = 0
      temp_w = w+'#'
      while (i<len(temp_w) and j<len(miss_word)):
        if miss_word[j]==temp_w[i]:
          ins_len += 1
          j+=1
          i+=1
        else:
          j+=1

      if ins_len==len(w):
        if (w in all_words):
          candidate_words.append(w)

    if (len(miss_word)==(len(w)-1)): #delete
      #print("deleteeee")
      del_len = 0
      i = 0
      j = 0
      temp_miss = miss_word+'#'

      while (i<len(w) and j<len(temp_miss)):
        #print(temp_miss, w)
        if temp_miss[j]==w[i]:
          del_len += 1
          j+=1
          i+=1
        else:
          i+=1

      if del_len==len(w)-1:
        if (w in all_words):
          candidate_words.append(w)

  print(candidate_words)
  return candidate_words
````

**P(x|w)**

````python
def prob(cand_words, miss_word, ins_table, del_table, sub_table, trans_table, bigrams):
  prob = {}

  if len(cand_words)==0:
    print("There are no candidate words")
    return prob

  for cand in cand_words:
    if len(miss_word)==len(cand): #substitute and transpose
      sub_len = 0
      trans_len = 0
      for i in range(0, len(miss_word)):
        if miss_word[i]==cand[i]:
          sub_len += 1
          trans_len += 1

      if (sub_len)==(len(cand)-1):
        for j in range(0, len(cand)):
          if miss_word[j] != cand[j]:
            row = ord(miss_word[j]) - 96
            col = ord(cand[j]) - 96
            numerator = sub_table[row][col]
            val = 0
            for k in range(0, 27):
              val+=sub_table[k][col]
            denominator = val
            prob[cand] = numerator/denominator

      if (trans_len)==(len(cand)-2):
        trans_flag = False
        i = 0
        while (i<len(miss_word)-1 and trans_flag == False):
          if (cand[i]!=miss_word[i]):
            if ((cand[i]==miss_word[i+1]) and (cand[i+1]==miss_word[i])):
              trans_flag = True
          i+=1

        if trans_flag == True:
          for j in range(0, len(cand)-1):
            if miss_word[j] != cand[j]:
              if ((cand[j]==miss_word[j+1]) and (cand[j+1]==miss_word[j])):
                row = ord(cand[j]) - 96
                col = ord(miss_word[j]) - 96
                numerator = trans_table[row][col]
                denominator = bigrams[cand[j]+miss_word[j]]
                prob[cand] = numerator/denominator
        
    if (len(miss_word)==(len(cand)+1)): #insert
      ins_len = 0
      i = 1
      j = 0
      temp_cand = '@'+cand
      while (i<len(temp_cand) and j<len(miss_word)):
        if miss_word[j]==temp_cand[i]:
          ins_len += 1
          j+=1
          i+=1
        else:
          j+=1

      if ins_len==len(cand):
        for j in range(0, len(temp_cand)):
          if (temp_cand[j] != miss_word[j]):
            if (temp_cand[j-1]=='@'):
              row = 0
              col = ord(miss_word[j]) - 96
            else:
              row = ord(temp_cand[j-1]) - 96
              col = ord(miss_word[j]) - 96
            numerator = ins_table[row][col]
            val = 0
            for k in range(0, 27):
              val+=sub_table[k][col]
            denominator = val
            prob[cand] = numerator/denominator

    if (len(miss_word)==(len(cand)-1)): #delete
      del_len = 0
      i = 0
      j = 0
      temp_miss = '@' + miss_word + '#'
      temp_cand = '@' + cand

      while (i<len(temp_cand) and j<len(temp_miss)):
        if temp_miss[j]==temp_cand[i]:
          del_len += 1
          j+=1
          i+=1
        else:
          i+=1

      if del_len==(len(temp_cand)-1):
        for j in range(0, len(temp_cand)):
          if (temp_miss[j] != temp_cand[j]):
            if (temp_miss[j-1]=='@'):
              row = 0
              col = ord(temp_cand[j]) - 96
            else:
              row = ord(temp_miss[j-1]) - 96
              col = ord(temp_cand[j]) - 96
            numerator = del_table[row][col]
            denominator = bigrams[temp_cand[j]+temp_miss[j]]
            prob[cand] = numerator/denominator

  return prob
````

# Finding the Most Probable Word

````python
def finding_most_probable(prob_cand_words, unigrams):
  if len(prob_cand_words) == 0:
    print("There is no most probable word since there are no candidate words")
    return ""
  total_prob = {}
  for w in prob_cand_words:
    if w in unigrams:
      total_prob[w] = prob_cand_words[w]*unigrams[w]
    else:
      print(w, "does not exist in vocabulary V")

  k = list(total_prob.keys())
  v = list(total_prob.values())
  max_word = k[0]
  max_value = v[0]
  for i in range(1, len(v)):
    if v[i] > max_value:
      max_value = v[i]
      max_word = k[i]

  return max_word
````
